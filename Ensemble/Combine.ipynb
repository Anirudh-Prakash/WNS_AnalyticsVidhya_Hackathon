```python
import pandas as pd
from sklearn.metrics import f1_score
```

```python
lgb=pd.read_csv('LGBM_Submission.csv').drop(['employee_id'],axis=1)
cb=pd.read_csv('CB_Submission.csv').drop(['employee_id'],axis=1)

new=pd.concat([lgb , cb], axis=1)
new.columns=['lgb', 'cb']
new.head()
```

```python
new.corr()
```

```python
new['final']= 0.4*new['lgb']+ 0.6*new['cb']
new.head(20)
```

```python
new['final']=new['final'].apply(lambda x:1 if x>0.28 else 0)
```

```python
new['final'].value_counts()
```

```python
test=pd.read_csv('test.csv')
Submission = pd.DataFrame({ 'employee_id': test['employee_id'],
                            'is_promoted':new['final']})
Submission.to_csv("Final_Submission.csv", index=False)
```
